```python
def count_short(row):
    sum_short=row.str.count("five").sum()
    return sum_short

def count_full(row):
    sum_full=row.str.count("nine").sum()
    return sum_full

#apply on axis=1 means you apply it on all rows (axis=0 would be column)
df["Short"] = np.nan
df['Full'] = np.nan
df['Short'] = df.apply(count_short, axis=1)
df['Full'] = df.apply(count_full, axis=1)


df
        
```

```python
def short_and_full(row):
    #print (row)
    if row.str.count("five").sum() != 0:
        if row.str.count("nine").sum() != 0:
            return 'in both'

df['both'] = df.apply(short_and_full, axis=1)
df
```
