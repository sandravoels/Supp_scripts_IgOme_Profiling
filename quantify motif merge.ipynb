```python
#open excel file with pd.read_excel()
df= pd.read_csv('sorted_cluster_to_combine400.csv')
df
```

```python
#generate new dataframe that removes all rows with have NaN in column "Motif 2"
df = df[df['Motif 2'].notna()]
df
```

```python
list_of_samples = ['141','269','457', '476','537','596','763','957']
```

```python
def find_sample(row):
    samples=[]
    for sample in list_of_samples:
        sum_full=row.str.count(sample).sum()
        if sum_full >=1:
            samples.append(sample)
    return samples

df['Sample'] = df.apply(find_sample, axis=1)
df
```

```python
def number_of_samples(field):
    return len(field)

df['Number of Samples'] = df['Sample'].apply(number_of_samples)
df
```
